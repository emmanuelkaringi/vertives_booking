```python
# split the hotel address and pick out the last string which would be the country.
# store the countries in a new column
data["cities"] = data.Hotel_Address.apply(lambda x: x.split(' ')[-1])
```

### Choose the required columns for the recommender

```python
# check available columns
data.columns
```

```python
# drop unecessary columns
data.drop(['Additional_Number_of_Scoring', 'Review_Date', 'Reviewer_Nationality', 'Negative_Review', 'Review_Total_Negative_Word_Counts', 'Total_Number_of_Reviews', 'Positive_Review', 'Review_Total_Positive_Word_Counts', 'Total_Number_of_Reviews_Reviewer_Has_Given', 'Reviewer_Score', 'days_since_review', 'lat', 'lng'],1, inplace=True)
```

```python
# create a function to convert the strings of list into a normal list
# get rid of the punctuations
def impute(column):
    column = column[0]
    if (type(column) != list):
        return "".join(literal_eval(column))
    else:
        return column
```

```python
# check on Tags column
data.head()
```

```python
# apply impute function to the Tags column to get rid of punctiation
data["Tags"] = data[["Tags"]].apply(impute, axis=1)
data.head()
```

```python
# convert Tags and Countries to lowercase
data["Tags"] = data["Tags"].str.lower()
data["cities"] = data["cities"].str.lower()
```

```python
# define the recommender function
def recommender(location, description):
    # devide the text into small tokens (sentences into words)
    description = description.lower()
    word_tokenize(description)
    #apply stopwords
    stop_words = stopwords.words('english')
    # apply lemmatization
    lemm = WordNetLemmatizer()
    # connect packages and filtering
    filtered = {word for word in description if not word in stop_words}
    filtered_set = set()
    for fs in filtered:
        filtered_set.add(lemm.lemmatize(fs))

# create variable that takes in the location and returns the following features
        city = data[data['cities']==location.lower()]
        city = city.set_index(np.arange(city.shape[0]))
        list1 = []; list2 = []; cos = [];
    for i in range(city.shape[0]):
        temp_token = word_tokenize(city["Tags"][i])
        temp_set = [word for word in temp_token if not word in stop_words]
        temp2_set = set()
        for s in temp_set:
            temp2_set.add(lemm.lemmatize(s))
        vector = temp2_set.intersection(filtered_set)
        cos.append(len(vector))
    # apply cosine similarity
    city['similarity'] = cos
    city = city.sort_values(by='similarity', ascending=False)
    city.drop_duplicates(subset='Hotel_Name', keep='first', inplace=True)
    city.sort_values('Average_Score', ascending=False, inplace=True)
    city.reset_index(inplace=True)
    return city[["Hotel_Name", "Average_Score", "Hotel_Address"]].head(10)
```

```python
# enter input
recommender('Kisumu', 'Business Trip')
```
